```python
# write your code here for VerificationType
# 0 Not set 
# 1 Income unverified 
# 2 Income unverified, cross-referenced by phone 
# 3 Income verified 
# 4 Income and expenses verified       

Verification_Type = {0 : "Not Set", 1 : "Income Unverified", 2 : "Income Unverified, Cross-Referenced By Phone", 3 : "Income Unverified", 4 : "Income And Expenses Verified"}

BondoraNew['VerificationType'] = BondoraNew['VerificationType'].apply(lambda x: Verification_Type.get(x))
BondoraNew['VerificationType'].value_counts()
```

```python
# write your code here Gender
BondoraNew["Gender"].value_counts()
```

```python
# write your code here
# 0  Male
# 1  Female
# 2 Undefined 

gen_type = {0 : "Male", 1 : "Female", 2 : "Undefined"}
BondoraNew["Gender"] = BondoraNew["Gender"].apply(lambda x: gen_type.get(x))

BondoraNew.Gender = BondoraNew.Gender.astype(str)
BondoraNew["Gender"].value_counts()
```

```python
# write your code here LanguageCode
BondoraNew["LanguageCode"].value_counts()
```

* As we can see from above in language code w ehave only descriptions for values 1,2,3,4,5,6, and 9 but it has other values too like 21,22,15,13,10 and 7 but they are very less it may happen they are local language codes whose decription is not present so we will be treated all these values as others

```python
# write your code here

# 1 Estonian 
# 2 English 
# 3 Russian 
# 4 Finnish 
# 5 German 
# 6 Spanish 
# 9 Slovakian

lang_code = {1 : "Estonian", 2 : "English", 3 : "Russian", 4 : "Finnish", 5 : "German", 6 : "Spanish", 9 : "Slovakian"}

BondoraNew["LanguageCode"] = BondoraNew["LanguageCode"].apply(lambda x: "Undefined" if (x not in lang_code) else lang_code.get(x))
BondoraNew["LanguageCode"].value_counts()
```

```python
# write your code here for UseOfLoan
BondoraNew["UseOfLoan"].value_counts()
```

* As we can see from above stats most of the loans are -1 category whose description is not avaialble in Bondoro website so we have dig deeper to find that in Bondora most of the loans happened for which purpose so we find in Bondora Statistics Page most of the loans around 34.81% are for Not set purpose. so we will encode -1 as Not set categor

```python
# write your code here
# 0 Loan consolidation 
# 1 Real estate 
# 2 Home improvement 
# 3 Business 
# 4 Education 
# 5 Travel 
# 6 Vehicle 
# 7 Other 
# 8 Health 
# 101 Working capital financing 
# 102 Purchase of machinery equipment 
# 103 Renovation of real estate 
# 104 Accounts receivable financing 
# 105 Acquisition of means of transport 
# 106 Construction finance 
# 107 Acquisition of stocks 
# 108 Acquisition of real estate 
# 109 Guaranteeing obligation 
# 110 Other business 
# All codes in format 1XX are for business loans that are not supported since October 2012

loan_use = {-1 : "Not Set", 0 : "Loan Consolidation", 1 : "Real Estate", 2 : "Home Improvement", 3 : "Business", 4 : "Education", 5 : "Travel", 6 : "Vehicle", 7 : "Other", 8 : "Health",
            101 : "Working Capital Financing", 102 : "Purchase Of Machinery Equipment", 103 : "Renovation of Real Estate", 104 : "Accounts Receivable Financing", 105 : "Aqquisition of Means of Transport",
            106 : "Construction Finance", 107 : "Aqquisition of Stocks", 108 : "Aqquisition of Means Of Transport", 109 : "Guarenteeing Obligation", 110 : "Other Business"}

BondoraNew["UseOfLoan"] = BondoraNew["UseOfLoan"].apply(lambda x: loan_use.get(x))
BondoraNew["UseOfLoan"].value_counts()
```

```python
# write your code here for Education
BondoraNew["Education"].value_counts()
```

* Again as we can see from above description for -1 and 0 in case of education is not present so we will encode them as Not_present as we dont know anything about them.

```python
# write your code here
# 1 Primary education 
# 2 Basic education 
# 3 Vocational education 
# 4 Secondary education 
# 5 Higher education

edu_type = {1 : "Primary Education", 2 : "Basic Education", 3 : "Vocational Education", 4 : "Secondary Education", 5 : "Higher Education"}
BondoraNew["Education"] = BondoraNew["Education"].apply(lambda x: "Not Present" if x not in edu_type else edu_type.get(x))

BondoraNew["Education"].value_counts()
```
